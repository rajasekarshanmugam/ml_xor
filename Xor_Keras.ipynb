```python
import numpy as np
print("np:", np.__version__)
```

```python
x = np.array([ [0., 0.], [0., 1.], [1., 0.], [1., 1.] ])
y = np.array([ 0., 1., 1., 0. ])
```

```python
import keras
from keras import layers
```

```python
model = keras.Sequential(
[
    layers.Input(shape=(2,)),
    layers.Dense(6, activation='relu'),
    layers.Dense(1)
])

model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.01), loss="mse", metrics=["accuracy"])
```

```python
es = keras.callbacks.EarlyStopping(monitor='loss', mode='min', min_delta=0.001, patience=3)
model.fit(x, y, batch_size=1, epochs=1000, verbose=1, callbacks=[es])
```

```python
y_hat = np.around(model.predict(x).reshape(-1))
y_true = np.reshape(y, -1)
print('expected:', y_true)
print('predicted:', y_hat)
```
